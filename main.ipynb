```python
import ollama


prompt = f"""
Extract the following details from this criminal judgment.

Return ONLY valid JSON.

Fields:
- prisoner_name
- case_number
- conviction_date
- judge_name
- court_name
- offence
- section
- sentence
- jail_name

Judgment:
{text[:15000]}
"""
response = ollama.chat(
    model='qwen3:8b',
    messages=[
        {
            'role': 'user',
            'content': prompt
        }
    ]
)

print(response['message']['content'])
```
